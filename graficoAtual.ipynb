```python
amostra1 = loadtxt('./controllers/walk/dados/Episodio.dat')
amostra2 = loadtxt('./controllers/walk2/dados/Episodio.dat')
amostra3 = loadtxt('./controllers/walk3/dados/Episodio.dat')
a = len(amostra1)
b = len(amostra2)
c = len(amostra3)
d = np.array([a, b, c])
print(np.min(d))
amostra4 = np.array([range(0, np.min(d)),range(0, np.min(d))])
amostra4 = amostra4.T
```

```python
for x in range(min(d)):
    amostra4[x][1] = (amostra1[x][1] + amostra2[x][1] + amostra3[x][1])/3
    #print(amostra1[x][1])
#plt.figure(figsize=(12, 12)) 
#plt.plot(amostra1)
#plt.savefig('./episodio4.svg')
```

```python
def movingaverage(interval, window_size):
    window= numpy.ones(int(window_size))/float(window_size)
    return numpy.convolve(interval, window, 'same')
```

```python
x = amostra4[:,0]
y = amostra4[:,1]

print(y.max())

plt.figure(figsize=(12, 12))
plt.plot(x,y,"k.", label='Recompensa Acumulada por episodio')
plt.legend()
y_av = movingaverage(y, 50)
plt.plot(x, y_av,"r", label='Media movel')
plt.legend()
xlim(0,len(amostra4))
plt.xlabel('Episodios', size=16)
plt.ylabel('Recompensa acumulada', size=16)
plt.grid(True)
#show()
```

```python
amostra1 = loadtxt('./controllers/walk/dados/Error.dat')
amostra2 = loadtxt('./controllers/walk2/dados/Error.dat')
amostra3 = loadtxt('./controllers/walk3/dados/Error.dat')
print(amostra1.max(),amostra2.max(),amostra3.max())
a = len(amostra1)
b = len(amostra2)
c = len(amostra3)
d = np.array([a, b, c])
print(np.min(d))
amostra5 = np.array([range(0, np.min(d)),range(0, np.min(d))])
amostra5 = amostra5.T
for x in range(min(d)):
    amostra5[x][1] = (amostra1[x][1] + amostra2[x][1] + amostra3[x][1])/3
```

```python
x = amostra5[:,0]
y = amostra5[:,1]

print(y.max())

plt.figure(figsize=(12, 8))
plt.plot(x,y,"k.", label='Erro do episodio')
plt.legend()
y_av = movingaverage(y, 50)
plt.plot(x, y_av,"r", label='Media movel')
plt.legend()
ylim(0,400000)
xlim(0,np.min(d))
plt.xlabel('Episodio', size=16)
plt.ylabel('$\sum_{t=0}^n({V_t - V_{t-1}})^2$', size=20)
plt.title('Curva de Convergencia', size=20)
plt.grid(True)
#show()
```
